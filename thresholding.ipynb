```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load orginal image
org_image=cv2.imread("images.jpg",0)
```

```python
#noise removal
removed_noise=cv2.medianBlur(org_image,5)
```

```python
plt.figure(figsize=[8,8])
plt.imshow(org_image,cmap="gray");plt.title("Orginal")
```

```python
_,output1=cv2.threshold(org_image,127,255,cv2.THRESH_BINARY)
plt.figure(figsize=[8,8])
plt.imshow(output1,cmap="gray");plt.title("Global Thresholding (v = 127)")
```

```python
output2=cv2.adaptiveThreshold(org_image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
plt.figure(figsize=[8,8])
plt.imshow(output2,cmap="gray");plt.title("Adaptive Mean Thresholding")
```

```python
output3=cv2.adaptiveThreshold(org_image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
plt.figure(figsize=[8,8])
plt.imshow(output3,cmap="gray");plt.title("Adaptive Gaussian Thresholding")
```

```python
_,output4 = cv2.threshold(removed_noise,127,255,cv2.THRESH_BINARY)
plt.figure(figsize=[8,8])
plt.imshow(output4,cmap="gray");plt.title("Adaptive Gaussian Thresholding")
```

```python
output5 = cv2.adaptiveThreshold(removed_noise,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,2)
plt.figure(figsize=[8,8])
plt.imshow(output5,cmap="gray");plt.title("Adaptive Gaussian Thresholding")
```

```python
output6 = cv2.adaptiveThreshold(removed_noise,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)
plt.figure(figsize=[8,8])
plt.imshow(output6,cmap="gray");plt.title("Adaptive Gaussian Thresholding")
```

```python
plt.figure(figsize=[15,15])
plt.subplot(331);plt.imshow(org_image, cmap='gray');plt.title("Original");
plt.subplot(332);plt.imshow(removed_noise, cmap='gray');plt.title("Noise removed");

plt.subplot(334);plt.imshow(output1, cmap='gray');plt.title("Global Thresholding (v = 127)");
plt.subplot(335);plt.imshow(output2, cmap='gray');plt.title("Adaptive Mean Thresholding");
plt.subplot(336);plt.imshow(output3, cmap='gray');plt.title("Adaptive Gaussian Thresholding");

plt.subplot(337);plt.imshow(output4, cmap='gray');plt.title("Global Thresholding + noise removal");
plt.subplot(338);plt.imshow(output5, cmap='gray');plt.title("Adaptive Mean + noise removal");
plt.subplot(339);plt.imshow(output6, cmap='gray');plt.title("Adaptive Gaussian + noise removal");
```
